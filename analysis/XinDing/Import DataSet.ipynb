```python
import pandas as pd
import gzip
import io
import requests

url='http://data.insideairbnb.com/united-states/nc/asheville/2021-04-19/data/listings.csv.gz'
response = requests.get(url)
bytes_io = io.BytesIO(response.content)
with gzip.open(bytes_io, 'rt') as read_file:
    df = pd.read_csv(read_file)
```
